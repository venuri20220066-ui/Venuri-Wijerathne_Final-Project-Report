```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/FYP_IIT/FYP_IIT_cleaned.csv', encoding='latin1')
df.head()
```

```python
import pandas as pd

# Factor 1
df['F1'] = df[
    ['curriculum_load','assignment_load','difficulty_catching_up',
     'exam_difficulty','exam_time_pressure','exam_stress']
].mean(axis=1)

# Factor 2
df['F2'] = df[
    ['module_failure','exam_concern','employment_concern']
].mean(axis=1)

# Factor 3
df['F3'] = df[
    ['academic_confidence','career_confidence','decision_making','time_management']
].mean(axis=1)

# Factor 4
df['F4'] = df[
    ['lecturer_strictness','lecturer_expectations']
].mean(axis=1)
```

```python
df.head()
```

```python
df = df[[
    'gender','academic_failure',
    'faculty_environment','home_environment',
    'age_group','academic_year','degree_program',
    'F1','F2','F3','F4'
]]
```

```python
df.head()
```

```python
df['age_group'] = df['age_group'].astype(str).str.strip()

df['age_group'] = df['age_group'].replace({
    '24-25': '24 and above',
    '24â25': '24 and above',
    'Above 25': '24 and above',
    '25': '24 and above',
    '25 above': '24 and above'
})
```

```python
df['age_group'] = df['age_group'].replace({
    '24-25': '24 and above',
    'Above 26': '24 and above'
})
```

```python
df['age_group'].value_counts()
```

```python
df['faculty_environment'] = df['faculty_environment'].astype(str).str.strip()

df['faculty_environment'] = df['faculty_environment'].replace({
    'Dissatisfied': 'Not satisfied',
    'Very dissatisfied': 'Not satisfied'
})
```

```python
df['faculty_environment'].value_counts()
```

```python
df.head()
```

```python
df['degree_program'] = df['degree_program'].replace({
    'B.B.A. (Hons) in Business Administration': 'Other',
    'B.B.A. (Hons) in Hospitality and Leisure Management': 'Other',
    'B.B.A. (Hons) in Human Resource Management': 'Other'
})
```

```python
df['degree_program'].value_counts()
```

```python
# Remove extra spaces and ensure string format
df['degree_program'] = df['degree_program'].astype(str).str.strip()

# Replace empty strings with 'General'
df['degree_program'] = df['degree_program'].replace('', 'General')

# Replace actual NaN values with 'General'
df['degree_program'] = df['degree_program'].replace('nan', 'General')
```

```python
df['degree_program'].value_counts()
```
